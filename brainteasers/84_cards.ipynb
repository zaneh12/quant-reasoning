# 84 Cards (IMC)

In a game, there is a deck of 84 distinct cards, each labeled 1 to 84. A dealer shuffles the deck and deals the top 5 cards. If they are in strictly ascending or strictly descending order, the player wins a payout of $x$; otherwise, they win nothing.

The game costs $1 to play.


### Question

What value of $x$ makes the game fair?

---

### Solution
> **Note:** The number 84 is a red herring. Since the cards are drawn uniformly at random, **only the relative order** of the 5 drawn cards matters â€” not their actual values. This would be identical if there were a million cards. The key is that **only 2 of the 120 possible orderings** are strictly increasing or decreasing.


## Solution

There are $5! = 120$ possible orderings of any 5-card hand.

Only 2 of these are strictly ordered: one ascending and one descending.

$$
P(\text{win}) = \frac{2}{5!} = \frac{1}{60}
$$

To break even:

$$
\frac{1}{60} \cdot x = 1 \Rightarrow x = 60
$$

Or more formally:

$$
\mathbb{E}[\text{profit}] = (\text{return} - \text{cost}) \cdot P(\text{win}) + (-\text{cost}) \cdot P(\text{lose}) = 0
$$

Where:
- $\text{return} = x$
- $\text{cost} = 1$
- $P(\text{win}) = \frac{1}{60}$
- $P(\text{lose}) = \frac{59}{60}$

Solving gives:

$$
x = 60
$$


---

### Final Answer: $\boxed{60}$
